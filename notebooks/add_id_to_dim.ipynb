# Ajout des ID numériques au table DIMENSION dans seeds

```python
import pandas as pd
import numpy as np
import shutil
```

```python
seeds_dir = '../snowflake/DBT/Projet_DBT/seeds'
src_data = '../pipeline/src/data'
```

```python
shutil.copytree(seeds_dir,'src_data',dirs_exist_ok=True)
```

### DIM_DOMAINES

```python
df_domaines = pd.read_csv(src_data+"/DIM_DOMAINE.csv")
df_domaines = df_domaines.reset_index(names='ID')
df_domaines = df_domaines.set_index('ID')
```

```python
df_domaines
```

```python
df_domaines.to_csv(src_data+'/DIM_DOMAINES.csv',encoding='utf-8')
```
